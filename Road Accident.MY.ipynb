```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots(2,3,figsize=(25,15))
sns.countplot(x='Urban_or_Rural_Area', data=df, ax=ax[0,0])
sns.countplot(x='Day_of_Week', data=df, ax=ax[0,1])
sns.countplot(x='Accident_Severity', data=df, ax=ax[0,2])
sns.countplot(x='Light_Conditions', data=df, ax=ax[1,0])
ax[1,0].tick_params(axis='x', rotation=90)
sns.countplot(x='Road_Type', data=df, ax=ax[1,1])
ax[1,1].tick_params(axis='x', rotation=90)
sns.countplot(x='Light_Conditions', data=df, ax=ax[1,2])
ax[1,2].tick_params(axis='x', rotation=90)
sns.countplot(x='Speed_limit', data=df, ax=ax[1,0])
```

```python
pivot_table = df.pivot_table(index=['Accident_Severity'], columns='Vehicle_Type', aggfunc='size', fill_value=0)
```

```python
pivot_table.plot(kind='bar', stacked=True)
plt.xlabel('Vehicle_Type & Accident_Severity')
plt.ylabel('Count')
plt.title('Vehicle_Type By Accident_Severity')
plt.legend(title='Accident_Severity')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Group by 'job_category' and calculate mean salary
demodf = df.groupby('Accident_Severity')['Speed_limit'].mean().reset_index()

# Sort the dataframe by 'salary_in_usd' in descending order and take the top 10
demodf = demodf.sort_values(by='Speed_limit', ascending=False).head(10)

# Set a custom color palette
custom_palette = sns.color_palette("viridis", n_colors=len(demodf))

# Bar Plot
plt.figure(figsize=(8, 5))
sns.barplot(y='Speed_limit', x='Accident_Severity', data=demodf, palette=custom_palette)
plt.xticks(rotation=90)
plt.title('Accident Severity By Speed Limit')
plt.ylabel('Speed_limit')
plt.xlabel('Accident_Severity')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style
sns.set(style="whitegrid")

# Create a countplot with hue and Set2 color palette
plt.figure(figsize=(10, 6))
plot1 = sns.countplot(x='Speed_limit', hue='Urban_or_Rural_Area', data=df, palette='Set2')

# Add labels and title
plt.title('Area vs Speed Limit by Days of the Week')
plt.xlabel('Area (Urban or Not)')
plt.ylabel('Count')

# Add legend
plt.legend(title='Speed Limit', loc='upper right')

# Remove spines for cleaner look
sns.despine()

# Show the plot
plt.show()
```
